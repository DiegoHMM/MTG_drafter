```python
import pandas as pd
```

```python
# Pool_card   -> # de Cartas que o usuário ja selecionou
# Pack_card   -> # de Cartas que saíram no pacote em questão
# Pack_Number -> # Rodada do pack atual
# Pick_Number -> # Rodada do pick atual
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("draft_data_public_BRO_PremierDraft.csv", chunksize=10000)
```

```python
for chunk in df:
    aux = chunk
    display(chunk)
    break
```

```python
all_col = []
```

```python

for column in aux.columns.unique():
    print(column)
    if 'pack_card_' in column:
        c = column.replace('pack_card_','')
        all_col.append(column)
    elif 'pool_' in column:
        c = column.replace('pool_','')
        all_col.append(column)
    
```
